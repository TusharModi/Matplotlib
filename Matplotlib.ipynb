```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='d')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',marker='d')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='d',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',marker='d')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='d',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',marker='d',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='d',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',marker='d',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend()
plt.grid()
```
